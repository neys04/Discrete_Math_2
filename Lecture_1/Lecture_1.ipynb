4. Prove Sedrakayan's Lemma $\forall u_i, v_i \in \mathbb{R}^+$:
\begin{gather}
\frac{(\Sigma_{i=1}^n u_i)^2}{\Sigma_{i=1}^n v_i}
\leq
\Sigma_{i=1}^n \frac{(u_i)^2}{v_i}
\end{gather}

**PROOF:**

Recall the Cauchy-Schwarz Inequality:

$$
|<a,b>|^{2} \leq\space\space <a,a><b,b>
$$

$$
\forall u_i, v_i \in \mathbb{R}^{+}
$$

Let $a_i = u_i, b_i = \sqrt{v_i}$

$$
|<a,b>|^{2}=(\sum_{i = 1}^{n}a_ib_i)^{2} = (\sum_{i = 1}^{n}u_i \sqrt{v_i})^{2}\\
$$
$$
<a,a><b,b>=(\sum_{i = 1}^{n}a_i^{2})(\sum_{i = 1}^{n}b_i^{2}) = (\sum_{i = 1}^{n} u_i^{2})(\sum_{i = 1}^{n}v_i)\\
$$

Thus,

$$
(\sum_{i = 1}^{n}u_i \sqrt{v_i})^{2} \leq (\sum_{i = 1}^{n} u_i^{2})(\sum_{i = 1}^{n}v_i)
$$

Dividing both sides by $\sum_{i=1}^{n}v_i$, we obtain:

$$
\frac{(\sum_{i=1}^{n}u_i)^{2}}{\sum_{i=1}^{n}v_i} \leq \sum_{i = 1}^{n} \frac{u_i^{2}}{v_i}
$$

Hence, Sedrakayan's Lemma is proven.

$$
\sqrt{(2x+1)^{2}} \geq \sqrt{|(2x + 1)+(2y - 1)|^{2}} - \sqrt{(2y-1)^{2}}

$$
\sqrt{(x)^{2}} = |x|
$$

Therefore, we get

$$
|2x + 1| \geq |(2x + 1) + (2y - 1)| - |2y - 1|
$$

Let us assign variables to simplify the expression, such that $a = 2x + 1$ and $b = 2y - 1$. Then we rewrite the inequality as:

$$
|a| \geq |a + b| - |b|
$$

By adding |b| and subtracting |a| to both sides and we get:

$$
|a + b| \leq |a| + |b|
$$

This inequality is known as the triangle inequality theorem.

Thus, the given inequality follows from the triangle inequality theorem.
